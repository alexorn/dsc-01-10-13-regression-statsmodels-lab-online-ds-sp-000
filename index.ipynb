### Step 1: Read the dataset and inspect its columns and 5-point statistics

```python
# Load necessary libraries and import the data
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
plt.style.use('seaborn')

adv_df = pd.read_csv('Advertising.csv',index_col =0)
```

```python
# Check the columns and first few rows
adv_df.head()
```

```python
# Get the 5-point statistics for data 
adv_df.describe()
```

```python
# Describe the contents of this dataset
# amount of sales and the amount of advertising in tv, radio, and newspaper
```

### Step 4: Run a simple regression in `statsmodels` with TV as a predictor

```python
# import libraries

# build the formula 
f1 = 'sales~TV'

# create a fitted model in one line
model = ols(formula = f1, data = adv_df).fit()
```

### Step 5: Get regression diagnostics summary


```python
model.summary()
```

### Record your observations on "Goodness of fit"


Note here that the coefficients represent associations, not causations


```python
#R-squared = 61% - relatively strong correlation, 61% of variance in sales can be explained using the spendings on TV
#slope - 0.05 (not a steep slope) - takes a lot of TV adv to generate sales (unless the units are different)
#p-val is 0; so there is a statistically significant relationship b/w sales and TV adv.
#skew = barely any skew, so normal distribution
#kurtosis = close to 3, so limited kurtosis in either direction
```

### Step 6:  Draw a prediction line with data points omn a scatter plot for X (TV) and Y (Sales)

Hint: We can use `model.predict()` functions to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

# create a DataFrame with the minimum and maximum values of TV


# make predictions for those x values and store them


# first, plot the observed data and the least squares line

```python
import numpy as np

X_new = pd.DataFrame ({'TV': [adv_df.TV.max(),adv_df.TV.min()]})
#maxy = model.predict(adv_df.TV.max())
#miny = model.predict(adv_df.TV.min())

preds = model.predict(X_new)
```

```python
adv_df.plot (kind = 'scatter',x='TV',y='sales')
plt.plot(X_new,preds, color = 'red')
plt.show()
```

### Step 7: Visualize the error term for variance and heteroscedasticity

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "TV", fig=fig)
plt.show()
```

```python
# Record Your observations on residuals
#y vs. x graph - variables are correlated and generally fit within the confidence levels
#residuals vs. TV = somewhat heteroscedasctity. residuals are not normally distributed. this breaks the assumption
#partial regression / CCPR - generally positively correlated (no additional independent variables yet)
```

### Next, repeat above for radio and go through the same process, recording your observations

```python
f2 = 'sales~radio'
model2 = ols(formula = f2, data = adv_df).fit()
X_new2 = pd.DataFrame ({'radio': [adv_df.radio.max(),adv_df.radio.min()]})
preds2 = model2.predict(X_new2)

adv_df.plot(kind = 'scatter',x='radio',y='sales')
plt.plot(X_new2,preds2,color='red')
plt.show()
```

```python
fig2 = plt.figure(figsize = (15,8))
fig2 = sm.graphics.plot_regress_exog(model2, "radio", fig=fig2)
plt.show()
```

```python
model2.summary()
```

```python
# Record your observations here for goodnes of fit 
# same as TV - heteroscedasticity


# newspaper - violated normality (normal distribution) and linearity (correlated)
# TV - violated heteroscdasticity (residuals are not normally distributed)
# radio - violated heteroscdasticity (residuals are not normally distributed)
```

## The Answer


Based on above analysis, we can conclude that none of the two chosen predictors is ideal for modeling a relationship with the sales volumes. `Newspaper` clearly violated normality and linearity assumptions. TV and radio did not provide a high value for co-efficient of determination - TV performed slightly better than the radio. There is obvious heteroscdasticity in the residuals for both variables.
